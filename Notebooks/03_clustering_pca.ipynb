
## **Clustering - PCA**

In this notebook, we will apply the **clustering** process, an **unsupervised learning** technique used to group similar objects within a dataset. The goal is to identify patterns and divide the data into clusters (groups) without the need for predefined labels.

To achieve optimal performance and high model accuracy, we will perform pre-processing, which includes techniques such as **dimensionality reduction** of the data.

```python
import pandas as pd

DATA = "../data/Customers_new_features_and_drop.csv"

df = pd.read_csv(DATA)

df.head()
```

```python
numerical_features = df.select_dtypes(include="number").columns
numerical_features
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(nrows=6, ncols=6, figsize=(14, 14), tight_layout=True)

for ax, feature in zip(ax.flatten(), numerical_features):
    sns.histplot(x=feature, ax=ax, data=df)
plt.show()
```

```python
one_hot_cols = [
    "Education",
    "Marital_Status",
    "AgeGroup",
    "HasChildren", 
    "HasAcceptedCmp", 
    "AcceptedCmp2", 
    "AcceptedCmp3", 
    "AcceptedCmp4", 
    "AcceptedCmp5", 
    "Complain", 
    "Response", 
    "Kidhome",
    "Teenhome",
    "YearsSinceEnrolled",
    "Children",
    "AcceptedCmpTotal"
]

standard_cols = [
    "Income",
    "Age"
]

power_cols = df.columns[
    df.columns.str.startswith("Mnt") | df.columns.str.startswith("Num")
].tolist()

minmax_cols = [
    col for col in df.columns if col not in one_hot_cols + standard_cols + power_cols
]
```

```python
from sklearn.preprocessing import  OneHotEncoder, StandardScaler, MinMaxScaler, PowerTransformer
from sklearn.compose import  ColumnTransformer

preprocessing = ColumnTransformer(
    [
        ("one-hot", OneHotEncoder(), one_hot_cols),
        ("standard", StandardScaler(), standard_cols),
        ("minmax", MinMaxScaler(), minmax_cols),
        ("power", PowerTransformer(), power_cols)
    ]
)

preprocessing.fit_transform(df)
```

```python
preprocessing.get_feature_names_out()
```

```python
df_scaled = pd.DataFrame(
    preprocessing.fit_transform(df),
    columns = preprocessing.get_feature_names_out()
)

df_scaled.head()
```


### **Defining the Number of Clusters**

The number of clusters to be used in the model is defined based on two methods:

- **Elbow Method** â†’ Evaluates the variability within the clusters (finds K by the "elbow").
- **Silhouette Method** â†’ Measures the cohesion and separation between clusters (finds K by the best score).


```python
from auxiliary_functions import plot_elbow_silhouette

plot_elbow_silhouette(df_scaled)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

RANDOM_STATE = 42

pipeline = Pipeline(
    [
        ("preprocessing", preprocessing),
        ("pca", PCA(n_components=2, random_state=RANDOM_STATE)),
        ("clustering", KMeans(n_clusters=3, random_state=RANDOM_STATE, n_init=10))
    ]
)

pipeline.fit(df)
```

```python
df_clustered = df.copy()
df_clustered["cluster"] = pipeline["clustering"].labels_

df_clustered.head()
```

```python
df_clustered["cluster"].value_counts().sort_index()
```
